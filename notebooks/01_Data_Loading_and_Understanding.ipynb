## Phase 2 : Feature Engineering


```python
#Extacting Year, Month, Quarter, and Day in different columns
# Year
df["Order Year"] = df["Order Date"].dt.year

# Month Number
df["Order Month"] = df["Order Date"].dt.month

# Month Name
df["Order Month Name"] = df["Order Date"].dt.month_name()

# Quarter
df["Order Quarter"] = df["Order Date"].dt.quarter

# Day Name
df["Order Day"] = df["Order Date"].dt.day_name()
```

```python
df.head()
```

```python
#Shipping Duration
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
df[['Order Date','Ship Date','Shipping Days']].head()
```

```python
#Profite Margin
df['Profit Margin(%)'] = (df['Profit']/df['Sales'])*100
```

```python
df[["Sales", "Profit", "Profit Margin(%)"]].head()
```

```python
#Dividing sales on the basis of low, medium, and high sales
df["Sales Category"] = pd.cut(df["Sales"], 
    bins=[0, 100, 500, float("inf")],
    labels=["Low", "Medium", "High"]
)
```

```python
df[["Sales", "Sales Category"]].head()
```

```python
#Similarly dividing Discount section into no discount, low, medium, and high discount
df['Discount Category'] = pd.cut(
    df['Discount'], bins=[-0.01, 0, 0.2, 0.5, 1],
    labels=['No Discount','Low','Medium','High']
)
```

```python
df[["Discount", "Discount Category"]].head()
```

```python
df["Order Status"] = np.where(df["Profit"] >= 0, "Profit", "Loss")
```

```python
df[["Sales", "Profit", "Order Status"]].head()
```

```python
#Saving cleaned dataset into data/procedded folder
df.to_csv("../data/processed/Superstore_Cleaned.csv", index=False)
```
