```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/Superstore_Cleaned.csv")
```

```python
df.columns
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_", regex=False)
      .str.replace("-", "_", regex=False)
      .str.replace("(", "", regex=False)
      .str.replace(")", "", regex=False)
      .str.replace("%", "pct", regex=False)
)
```

```python
df.columns.tolist()
```

```python
df.rename(columns={"profit_marginpct": "profit_margin_pct"},inplace=True)
```

```python
from sqlalchemy import create_engine

engine = create_engine(
    "mysql+pymysql://root:%4011ZAID@localhost/customer_retention_db"
)
```

```python
#Loading data into MySQL
df.to_sql(
    "superstore",
    con=engine,
    if_exists="replace",
    index=False
)
```
